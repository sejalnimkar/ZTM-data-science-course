```python

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot on the top-left subplot
axs[0, 0].plot([1, 2, 3], [4, 5, 6])

# Plot on the top-right subplot
axs[0, 1].scatter([1, 2, 3], [4, 5, 6])

# Plot on the bottom-left subplot
axs[1, 0].bar([1, 2, 3], [4, 5, 6])

# Display the figure with subplots
plt.show()
```

```python
# make a plot from dictionary 
nut_butter_prices={"almond":10,"peanut":8,"cashew":12}

fig,ax=plt.subplots()
ax.bar(x=list(nut_butter_prices.keys()),height=list(nut_butter_prices.values()))
ax.set_title('butter store')
ax.set_xlabel('nut butter')
ax.set_ylabel('Frequency')
plt.show()
```

```python
fig,ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))

plt.show()
```

```python
over50.plot(kind='scatter',x='age',y='chol',c='target')
```

```python
fig,axs=plt.subplots(figsize=(10,6))
over50.plot(kind='scatter',x='age',y='chol',c='target',ax=axs);

# axs.set_xlim([45,100])
```

```python
fig,axs=plt.subplots(2,1,figsize=(10,6),sharex=True)
scatter=axs[0].scatter(x=over50['age'],y=over50['chol'],c=over50['target']);

axs[0].set(title='Heart Disease and Cholestrol plot',ylabel='cholestrol level')
axs[0].legend(*scatter.legend_elements(),title='Target')

axs[0].axhline(over50['chol'].mean(),linestyle='--')

axs[1].scatter(x=over50['age'],y=over50['thalach'],c=over50['target'])
axs[1].set(title='Heart disease and max heart rate',xlabel='Age',ylabel='Max heart rate');
axs[1].legend(*scatter.legend_elements(),title='Target')
axs[1].axhline(y=over50['thalach'].mean(),linestyle='--')


fig.suptitle('Heart disease analysis',fontsize=16, fontweight='bold')
```
